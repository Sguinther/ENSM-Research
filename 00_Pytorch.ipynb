#Rules that need to be satisfied for matrix multiplication:

# 1. inner dimensions must match:
- (3,2) @ (3,2) wont work
- (2,3) @ (3,2) will work
- (3,2) @ (2,3) will work 

2. The resulting matrix has the shape of the outer dimensions:
- (2,3) @ (3,2) -> (2,2)
- (3,2) @ (2,3) -> (3,3)

```python
tensorAA = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])

tensorBB = torch.tensor([[7,10],
                        [8,11],
                        [9,12]])
# torch.mm(tensorAA, tensorBB) alias for torch.matmul
torch.matmul(tensorAA, tensorBB)

```

#To fix shape issues, we can manipulate the shape of our tensors using transpose

The transpose switches the axes or dimensions of a given tensor

```python
tensorAA = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])

tensorBB = torch.tensor([[7,10],
                        [8,11],
                        [9,12]])
# torch.mm(tensorAA, tensorBB) alias for torch.matmul
#torch.matmul(tensorAA, tensorBB)
print(tensorAA, tensorAA.shape, f"tensorA")

print(f"tensorB")
print(tensorBB.shape)
print(tensorBB)

#.T transposes the tensor

print(tensorBB.T.shape)

print(tensorBB.T)

output = torch.matmul(tensorAA, tensorBB.T)
print(output)
```

## Reshaping, stacking, squeezing and unsqueezing tensors

* reshaping - reshapes input tensor to a defined shape
* view - return a view of an input tensor of a certain shape but keep the same memory as the original tensor 
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack) https://pytorch.org/docs/stable/generated/torch.stack.html 
* Squeeze - removes all 1 dimensions from a tensor
* unsqueeze - add a 1 dimension to a tensor
* permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
# torch.squeeze() - removes all single dimensions from target tensor
tensor111 = torch.arange(1,13)
tensor11 = tensor111.reshape(1,12)
print(f"previous tensor: {tensor11}")
print(f"previous shape: {tensor11.shape}")

#remove extra dimensions from tensor11
squeezed11 = tensor11.squeeze()
print(f"new tensor: {squeezed11}")
print(f"previous shape: {squeezed11.shape}")

#Then torch.unsqueeze() to add dimension
newnew= squeezed11.unsqueeze(dim=1)
print("unsqueeze:", newnew, newnew.shape)



```

```python
#torch.permute - basically means rearange dimensions
xx = torch.rand(size = (224,224,3)) #height, width, color channels, image data standard

#permute original tensor to rearange the dimensions
xxx = xx.permute(2,0,1) # shifts dimensions 2 to 0, 0 to 1, and 1 to 2

print(f"previous shape: {xx.shape}")
print(f"new shape: {xxx.shape}")
```

##Reproducability (trying to take random out of random)

In short how a neural network learns:

Start with random numbers -> tensor operations -> update random numbers to try and make them  better representations of the data -> again -> again -> again.....

To reduce randomness in neural networks and PyTorch comes the concept of a random seed.

Essentially a random seed " flavors" the randomness

```python
import torch

#Two random tensors

randomA = torch.rand(3,4)
randomB = torch.rand(3,4)

print(randomA)
print(randomB)
print(randomA == randomB)
```

```python
# Lets make some random but reproducable tensors

import torch

#set random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
randomC = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
randomD = torch.rand(3,4)

print(randomC, '\n', randomD)
```
